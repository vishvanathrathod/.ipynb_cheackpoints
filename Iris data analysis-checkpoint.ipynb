```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 

import warnings
warnings.filterwarnings("ignore") 

sns.set_style("darkgrid")
```

```python
data = pd.read_csv('IRIS.csv')

data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data['species'].value_counts()
```

```python
sns.pairplot(data=data, hue='species',markers=["v", "<", ">"],height=5,corner=False)

plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15,15))
 
axes[0,0].set_title("Sepal Length\n",fontsize=15,color="green")
axes[0,0].hist(data['sepal_length'], bins=7)
 
axes[0,1].set_title("Sepal Width\n",fontsize=15,color="green")
axes[0,1].hist(data['sepal_width'], bins=5)
 
axes[1,0].set_title("Petal Length\n",color="red",fontsize=15)
axes[1,0].hist(data['petal_length'], bins=6)
 
axes[1,1].set_title("Petal Width\n",fontsize=15,color="red")
axes[1,1].hist(data['petal_width'], bins=6)
```

```python
g = sns.violinplot(y='species', x='sepal_length', data=data, inner='quartile')
plt.show()
g = sns.violinplot(y='species', x='sepal_width', data=data, inner='quartile')
plt.show()
g = sns.violinplot(y='species', x='petal_length', data=data, inner='quartile')
plt.show()
g = sns.violinplot(y='species', x='petal_width', data=data, inner='quartile')
plt.show()
```
